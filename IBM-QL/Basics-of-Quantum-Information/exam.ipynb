```python
# we can use the measure method on a compound system, also specifying 
# explicitly which qubit we want to measure

# np.array in here contains probabilities of each basis vector. 
# Eg. if we give 8 entries, it recognizes 3-qubit state and prints non-zero ones:

v1 = Statevector.from_label('+').tensor(Statevector.from_label('-'))
v1 = Statevector.from_label('0').tensor(v1)
display(v1.draw("latex"))

v2 = Statevector.from_label('0').tensor(Statevector.from_label('0'))
v2 = Statevector.from_label('1').tensor(v2)
display(v2.draw("latex"))

v3 = Statevector.from_label('1').tensor(Statevector.from_label('1'))
v3 = Statevector.from_label('1').tensor(v3)
display(v3.draw("latex"))

psi = (v1 + v2 + v3) / np.sqrt(3)
display(psi.draw("latex"))

result, state = psi.measure([1])
print(f"Measured: {result}\nState after measurement:")
display(state.draw("latex"))

# tmp = Statevector.from_label('0').tensor(Statevector.from_label('-'))
# tmp = Statevector.from_label('0').tensor(tmp)
# display(tmp.draw("latex"))

# tmp = Statevector(ket0 @ ket1)
# display(tmp.draw('latex'))
# print(ket0, ket1)
# print(np.matmul(ket0, ket1))

# # The actuall example 
# print("The actuall example:")
# w = Statevector([0, 1, 1, 0, 1, 0, 0, 0] / np.sqrt(3))
# display(w.draw("latex"))

# result, state = w.measure([0])
# print(f"Measured: {result}\nState after measurement:")
# display(state.draw("latex"))

# result, state = w.measure([0,1])
# print(f"Measured: {result}\nState after measurement:")
# display(state.draw("latex"))
```

### Q15

```python
psi_0 = Statevector(np.array([1/np.sqrt(3), np.sqrt(2/3)]))
psi_1 = Statevector(np.array([np.sqrt(2/3), -1/np.sqrt(3)]))

display(psi_0.draw('latex'))
display(psi_1.draw('latex'))

qsv = Statevector(np.array([np.sqrt(2)/3, -1/3, np.sqrt(2)/3, 2/3]))
display(qsv.draw('latex'))

pi_0 = np.outer(psi_0.data, np.conjugate(psi_0.data))
op_pi_0 = np.kron(np.eye(2), pi_0)
display(Operator(op_pi_0).draw('latex'))

exp_val = qsv.data.conj().T @ op_pi_0 @ qsv.data

print('probability of obtaining the outcome 0 = {:.3f}.'.format(exp_val.real))
```
